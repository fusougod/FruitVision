```python
from ultralytics import YOLO

# 1. Create a YOLOv8 model for classification
model = YOLO('yolov8n-cls.pt')  # Use a small pretrained YOLOv8 classification model

# 2. Train the model
model.train(
    data='D:\ComVision - Yolo\Fruits Classification',  # Path to the dataset directory
    epochs=20,               # Number of training epochs
    imgsz=128,               # Image size (resize all images to 128x128)
    batch=32,                # Batch size
    device='cpu'                 # Use GPU (if available), otherwise CPU
)

# 3. Validate the model on the validation set
metrics = model.val(data='D:\ComVision - Yolo\Fruits Classification')

# Save the trained model
model.save('yolov8_fruit_classifier.pt')
```

```python
from ultralytics import YOLO
import os

# Load the saved model
model = YOLO('yolov8_fruit_classifier.pt')

# Make predictions using the saved model
results = model.predict(source='D:/CompVision - Images for model testing/what in tarnation.webp')

# Specify the directory where you want to save the output image
output_dir = 'D:/CompVision - Images for model testing/output/'

# Ensure the output directory exists
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Find the next available counter for the output file
counter = 1
while os.path.exists(os.path.join(output_dir, f'apol_pred_{counter}.jpg')):
    counter += 1

# Create the file name with the counter
output_file = os.path.join(output_dir, f'apol_pred_{counter}.jpg')

# Save the result with bounding boxes plotted
results[0].save(filename=output_file)

# Optionally, display the image with bounding boxes
results[0].show()

# Print the results as well to check the details of the predictions
print(results)
```

```python
from ultralytics import YOLO
import cv2

# Load the saved model
model = YOLO('yolov8_fruit_classifier.pt')

# Load the image and resize it to 640x640 (or the input size expected by the model)
img_path = 'D:/CompVision - Images for model testing/what in tarnation.webp'
img = cv2.imread(img_path)
img_resized = cv2.resize(img, (640, 640))

# Make predictions using the resized image
results = model.predict(source=img_resized)

# Access the first result (as results is a list)
pred = results[0]

# Check if any boxes were detected
if pred.boxes is not None and len(pred.boxes) > 0:
    # Convert the image to a format compatible with OpenCV (numpy array)
    img = pred.orig_img  # Original image

    # Plot the bounding boxes
    for box in pred.boxes:
        # Get the box coordinates and label
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()  # Bounding box coordinates
        conf = box.conf[0].cpu().numpy()  # Confidence score
        cls = box.cls[0].cpu().numpy()  # Class ID
        label = f'{model.names[int(cls)]} {conf:.2f}'  # Class label and confidence

        # Only draw the box if the confidence is above the threshold
        if conf >= 0.3:
            # Draw the bounding box and label on the image
            cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)  # Draw box
            cv2.putText(img, label, (int(x1), int(y1)-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)  # Put label

    # Save the image with bounding boxes
    output_path = 'D:/CompVision - Images for model testing/what.png'

    print(f"Image saved to {output_path}")
else:
    print("No objects detected in the image.")

# Optionally, print the results as well
print(results)
```
